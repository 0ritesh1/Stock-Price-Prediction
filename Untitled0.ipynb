```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
file_path = '/content/stock_data.csv'
df = pd.read_csv(file_path)
```

```python
df.dropna(inplace=True)  # Remove missing values
df = df.select_dtypes(include=[np.number])
```

```python
# Exploratory Data Analysis (EDA)
plt.figure(figsize=(10, 5))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Feature Correlation')
plt.show()
```

```python
# Feature Selection and Scaling
features = df.iloc[:, :-1]  # All columns except last as features
target = df.iloc[:, -1]  # Last column as target
```

```python
scaler = MinMaxScaler()
features_scaled = scaler.fit_transform(features)
```

```python
# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(features_scaled, target, test_size=0.2, random_state=42)
```

```python
# Model Training (Random Forest Regressor)
model = RandomForestRegressor(n_estimators=200, random_state=42, max_depth=10)
model.fit(X_train, y_train)

```

```python
# Predictions
y_pred = model.predict(X_test)
```

```python
# Model Evaluation
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)
```

```python
print(f'MAE: {mae}')
print(f'MSE: {mse}')
print(f'RMSE: {rmse}')
print(f'R2 Score: {r2}')
```

```python
# Visualization
plt.figure(figsize=(10, 5))
plt.plot(y_test.values, label='Actual', marker='o', linestyle='dashed')
plt.plot(y_pred, label='Predicted', marker='x', linestyle='solid')
plt.xlabel('Test Samples')
plt.ylabel('Stock Price')
plt.legend()
plt.title('Stock Price Prediction')
plt.grid()
plt.show()
```

```python
# Feature Importance Visualization
importances = model.feature_importances_
feature_names = df.columns[:-1]
```

```python

plt.figure(figsize=(10, 5))
sns.barplot(x=importances, y=feature_names, palette='viridis')
plt.title('Feature Importance')
plt.xlabel('Importance Score')
plt.ylabel('Features')
plt.grid()
plt.show()
```
